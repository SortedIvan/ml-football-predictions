># **Plotting results**

```python
import matplotlib.pyplot as plt
#Goals scored by host teams 
# - sorted by scores and amount of games
matchGoals_df = matches_df['host_scores'].value_counts()
plotGoalsByHost = matchGoals_df.plot.pie(y='host_scores', figsize=(7, 7))
```

```python
matchGoals_df = matches_df['host_scores'].value_counts()
matchGoals_df.plot(kind="bar",figsize = (5,5));
```

```python
#Goals scored by guest teams 
# - sorted by scores and amount of games
matchGoals_df = matches_df['guest_scores'].value_counts()
plotGoalsByGuest = matchGoals_df.plot.pie(y='guest_scores', figsize=(7, 7))
```

```python
#Goals scored by guest teams 
# - sorted by scores and amount of games
matchGoals_df = matches_df['guest_scores'].value_counts()
matchGoals_df.plot(kind="bar",figsize = (5,5));
```

```python
#Amount of scored goals in every minute, grouped by amount of scored goals
goals_df.plot(kind="bar",figsize = (30,10));
```

```python
# Amount of scores from home and guest teams.
# We can clearly see an increase when players are playing in a familiar stadium/playground
goalsScoredByHomeTeam_df.sum()
goalsScoredByGuestTeam_df.sum()
goals1and2 = pd.DataFrame({'Home Goals': [goalsScoredByHomeTeam_df.sum(),goalsScoredByGuestTeam_df.sum()]},
                  index=['Home', 'Guest'])
plot = goals1and2.plot.pie(y='Home Goals', figsize=(5, 5))
```

```python
goalsScoredByHomeTeam_df = scoredGoals_df['player'].loc[scoredGoals_df['team'] == scoredGoals_df['host']].value_counts()
goalsScoredByGuestTeam_df = scoredGoals_df['player'].loc[scoredGoals_df['team'] == scoredGoals_df['guest']].value_counts()
goalsScoredByGuestTeam_df.sum()
goalsScoredByHomeTeam_df.sum()
df = pd.DataFrame({'Clubs':['Host', 'Guest'], '#_of_Goals':[goalsScoredByHomeTeam_df.sum(), goalsScoredByGuestTeam_df.sum()]})
df.plot.bar(x='Clubs', y='#_of_Goals', rot=0)
```

```python
# x axis displays the (first 20) amount of spectators in thousands, the y axis - the number of occurences
matchSpectators_df.head(20).plot(kind="bar",figsize = (30,10));
```

```python
hostingTeams_df = pd.DataFrame(hostingTeams_df)
hostingTeams_df.plot(kind="bar",figsize = (30,10));
```

```python
guestTeams_df = pd.DataFrame(guestTeams_df)
guestTeams_df.plot(kind="bar",figsize = (30,10));
```
